```python
import cv2
import os
from detectron2.config import get_cfg
from detectron2.engine import DefaultPredictor
from detectron2.utils.visualizer import Visualizer, ColorMode
from detectron2.data import MetadataCatalog
from detectron2.model_zoo import get_config_file

# === Configuration ===
cfg = get_cfg()
cfg.merge_from_file(get_config_file("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"))
cfg.MODEL.DEVICE = "cuda"  # use "cuda" if GPU is available
cfg.DATASETS.TEST = ("my_dataset_val",)
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 1  # filter, no filter
cfg.MODEL.WEIGHTS = "filter_model.pth"
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5

predictor = DefaultPredictor(cfg)
MetadataCatalog.get("my_dataset_val").thing_classes = ["filter", "no filter"]
metadata = MetadataCatalog.get("my_dataset_val")

# === ROI Coordinates ===
roi_value = {'x': 1359, 'y': 528, 'width': 327, 'height': 448}
x, y, w, h = roi_value['x'], roi_value['y'], roi_value['width'], roi_value['height']

# === Video Setup ===
video_path = "input.ts"
output_path = "output_video.mp4"

cap = cv2.VideoCapture(video_path)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# === Process Video Frame-by-Frame ===
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    roi = frame[y:y+h, x:x+w]
    outputs = predictor(roi)
    instances = outputs["instances"].to("cpu")
    pred_classes = instances.pred_classes.tolist()

    # === Classify Frame ===
    if 0 in pred_classes:
        label = "FILTER"
        color = (0, 255, 0)
    elif 1 in pred_classes:
        label = "NO FILTER"
        color = (0, 0, 255)
    else:
        label = "NO DETECTION"
        color = (128, 128, 128)

    # === Visualize and paste ROI back ===
    visualizer = Visualizer(roi[:, :, ::-1], metadata=metadata, scale=0.8, instance_mode=ColorMode.IMAGE)
    vis_roi = visualizer.draw_instance_predictions(instances)
    vis_roi_bgr = vis_roi.get_image()[:, :, ::-1]

    # Resize and paste
    vis_roi_resized = cv2.resize(vis_roi_bgr, (w, h))
    frame[y:y+h, x:x+w] = vis_roi_resized

    # Draw ROI rectangle and classification label
    cv2.rectangle(frame, (x, y), (x+w, y+h), color, 2)
    cv2.putText(frame, f"Prediction: {label}", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)

    out.write(frame)

# === Cleanup ===
cap.release()
out.release()
cv2.destroyAllWindows()

print("✅ Finished processing video with ROI-based classification and visualization.")
```

```python
import cv2
import time
from detectron2.config import get_cfg
from detectron2.engine import DefaultPredictor
from detectron2.utils.visualizer import Visualizer, ColorMode
from detectron2.data import MetadataCatalog
from detectron2.model_zoo import get_config_file

# === Step 1: Configuration ===
cfg = get_cfg()
cfg.merge_from_file(get_config_file("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"))

cfg.MODEL.WEIGHTS = "/content/model_final.pth"  # Trained model
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.9
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 1  # Only one class: "milk"
cfg.MODEL.DEVICE = "cuda"  # Use "cpu" if GPU is unavailable

# === Step 2: Register Metadata ===
DUMMY_DATASET_NAME = "milk_dataset"
MetadataCatalog.get(DUMMY_DATASET_NAME).thing_classes = ["milk"]
metadata = MetadataCatalog.get(DUMMY_DATASET_NAME)

# === Step 3: Create Predictor ===
predictor = DefaultPredictor(cfg)

# === Step 4: Video Input/Output Setup ===
input_video_path = "/content/video.mp4"
output_video_path = "/content/output_video_1.mp4"

cap = cv2.VideoCapture(input_video_path)
if not cap.isOpened():
    raise IOError("❌ Cannot open input video.")

fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Try 'avc1' or 'XVID' if needed
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

# Optional: Skip to a specific frame (e.g., frame 450)
# cap.set(cv2.CAP_PROP_POS_FRAMES, 450)

# === Step 5: Process Each Frame ===
frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Predict
    start = time.time()
    outputs = predictor(frame)
    end = time.time()

    # Visualize
    instances = outputs["instances"].to("cpu")
    visualizer = Visualizer(frame[:, :, ::-1], metadata=metadata, scale=1.0, instance_mode=ColorMode.IMAGE)
    vis_output = visualizer.draw_instance_predictions(instances)
    vis_frame = vis_output.get_image()[:, :, ::-1]  # Convert RGB → BGR

    # Check shape and type
    if vis_frame.shape == (height, width, 3) and vis_frame.dtype == 'uint8':
        out.write(vis_frame)
    else:
        print(f"⚠️ Skipping frame {frame_count}: invalid shape or type.")

    print(f"✅ Frame {frame_count} processed in {end - start:.2f}s")
    frame_count += 1

# === Step 6: Cleanup ===
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"🎉 Done! Processed {frame_count} frames. Video saved at {output_video_path}")
```
